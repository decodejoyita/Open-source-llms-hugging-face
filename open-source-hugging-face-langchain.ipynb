```python
from langchain_huggingface import HuggingFaceEndpoint
```

```python
from google.colab import userdata
sec_key=userdata.get("hf_token")
```

```python
import os
os.environ['HUGGINGFACEHUB_API_TOKEN'] = sec_key
```

```python
model_id="mistralai/Mistral-7B-Instruct-v0.3"
```

```python
llm=HuggingFaceEndpoint(repo_id=model_id, max_length=128, temperature=0.7, token=sec_key)
```

```python
llm.invoke("The key to life is")
```

```python
from langchain import PromptTemplate, LLMChain
```

```python
question="Who won the cricket world cup in 2011?"

template="""Question: {question}
Answer: Let's think step by step."""

prompt=PromptTemplate(template=template, input_variables=["question"])
```

```python
llm_chain=LLMChain(llm=llm,prompt=prompt)
print(llm_chain.invoke(question))
```

```python
from langchain_huggingface import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
model_id="gpt2"
model=AutoModelForCausalLM.from_pretrained(model_id)
tokenizer=AutoTokenizer.from_pretrained(model_id)
```

```python
pipe=pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=100)
hf=HuggingFacePipeline(pipeline=pipe)
```

```python
hf.invoke("What is machine learning?")
```

```python
gpu_llm=HuggingFacePipeline.from_model_id(model_id=model_id, task="text-generation", pipeline_kwargs={"max_new_tokens":100})
```
